```python
# ============================
# Full upgraded pipeline cell
# Paste this single cell in Colab (run the install step first)
# ============================

# INSTALL (run once)
!pip install -q transformers pillow matplotlib scikit-image torch torchvision psutil GPUtil tqdm

# IMPORTS
import time
import json
import os
import io
from datetime import datetime
from collections import Counter, defaultdict

import psutil
import GPUtil
from tqdm import tqdm

import torch
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from skimage.transform import resize
from skimage.filters import sobel

from transformers import BlipProcessor, BlipForConditionalGeneration
from torchvision import transforms as T
import torchvision
from google.colab import files

from matplotlib.backends.backend_pdf import PdfPages

# ---------------------------
# Configuration (tweak if needed)
# ---------------------------
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
DETECTION_THRESHOLD = 0.5
MAX_DISPLAY_DETECTIONS = 50
WARMUP_RUNS = 1             # warm-up passes to stabilize measured times
REPEATS_PER_IMAGE = 2       # number of times to run caption/detection for stats

OUTPUT_DIR = "/content/model_report_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------
# Safe utilities
# ---------------------------
def safe_makedirs(path):
    try:
        os.makedirs(path, exist_ok=True)
    except Exception:
        pass

def now_str():
    return datetime.utcnow().strftime("%Y%m%dT%H%M%SZ")

# ---------------------------
# Show utilities
# ---------------------------
def show_image(img, title=None, figsize=(6,6)):
    plt.figure(figsize=figsize)
    if isinstance(img, Image.Image):
        plt.imshow(np.array(img))
    else:
        plt.imshow(img)
    plt.axis('off')
    if title:
        plt.title(title)
    plt.show()

def overlay_heatmap_on_image(image_pil, heatmap, alpha=0.5):
    img = np.array(image_pil).astype(np.float32) / 255.0
    hm = np.array(heatmap).astype(np.float32)
    hm = hm - hm.min()
    if hm.max() > 0:
        hm = hm / hm.max()
    import matplotlib.cm as cm
    cmap = cm.get_cmap("jet")
    heat = cmap(hm)[:, :, :3]
    if heat.shape[:2] != img.shape[:2]:
        heat = resize(heat, img.shape[:2], order=1, mode='reflect', anti_aliasing=True)
    blended = (1 - alpha) * img + alpha * heat
    blended = np.clip(blended, 0, 1)
    return Image.fromarray((blended * 255).astype(np.uint8))

# ---------------------------
# Load models with exception handling (preserve workflow)
# ---------------------------
print("Loading models (this may take a moment)...")
processor = None
blip_model = None
det_model = None

try:
    processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(DEVICE)
    blip_model.eval()
    print("BLIP model loaded on", DEVICE)
except Exception as e:
    print("ERROR loading BLIP model:", e)

try:
    det_model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True).to(DEVICE)
    det_model.eval()
    print("Detection model loaded on", DEVICE)
except Exception as e:
    print("ERROR loading detection model:", e)

# ---------------------------
# Detection label map (safe lookup)
# ---------------------------
COCO = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'N/A', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'N/A', 'backpack', 'umbrella',
    'N/A', 'N/A', 'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard',
    'sports ball', 'kite', 'baseball bat', 'baseball glove', 'skateboard',
    'surfboard', 'tennis racket', 'bottle', 'N/A', 'wine glass', 'cup', 'fork',
    'knife', 'spoon', 'bowl', 'banana', 'apple', 'sandwich', 'orange', 'broccoli',
    'carrot', 'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch', 'potted plant',
    'bed', 'N/A', 'dining table', 'N/A', 'N/A', 'toilet', 'N/A', 'tv', 'laptop',
    'mouse', 'remote', 'keyboard', 'cell phone', 'microwave', 'oven', 'toaster',
    'sink', 'refrigerator', 'N/A', 'book', 'clock', 'vase', 'scissors', 'teddy bear',
    'hair drier', 'toothbrush'
]

def safe_label_lookup(label_index):
    li = int(label_index)
    if 0 <= li < len(COCO):
        name = COCO[li]
        if name is None or name == "N/A":
            return f"label_{li}"
        return name
    else:
        return f"label_{li}"

# ---------------------------
# Inference functions (wrapped, unchanged workflow)
# ---------------------------
to_tensor = T.Compose([T.ToTensor()])

def generate_caption(image, max_length=40, num_beams=3):
    """
    Returns caption and time_taken (seconds). Handles missing BLIP gracefully.
    """
    if blip_model is None or processor is None:
        return "Caption model not available", None
    inputs = processor(images=image, return_tensors="pt").to(DEVICE)
    t0 = time.time()
    with torch.no_grad():
        out = blip_model.generate(**inputs, max_length=max_length, num_beams=num_beams)
    t1 = time.time()
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption, (t1 - t0)

def run_detection(image, threshold=DETECTION_THRESHOLD, max_boxes=MAX_DISPLAY_DETECTIONS):
    """
    Returns list of (bbox, label_str, score), and time_taken (seconds).
    Uses optimized detach & cpu once.
    """
    if det_model is None:
        return [], None
    img_t = to_tensor(image).to(DEVICE)
    t0 = time.time()
    with torch.no_grad():
        outputs = det_model([img_t])
    t1 = time.time()
    if not outputs or len(outputs) == 0:
        return [], (t1 - t0)
    preds = outputs[0]
    boxes = preds.get("boxes")
    labels = preds.get("labels")
    scores = preds.get("scores")
    if boxes is None or labels is None or scores is None:
        return [], (t1 - t0)
    boxes_cpu = boxes.detach().cpu().numpy()
    labels_cpu = labels.detach().cpu().numpy()
    scores_cpu = scores.detach().cpu().numpy()
    keep = np.where(scores_cpu >= threshold)[0]
    if keep.size == 0:
        return [], (t1 - t0)
    order = keep[np.argsort(-scores_cpu[keep])]
    if max_boxes is not None:
        order = order[:max_boxes]
    results = []
    for idx in order:
        b = boxes_cpu[idx]
        l = labels_cpu[idx]
        s = float(scores_cpu[idx])
        results.append((b, safe_label_lookup(l), s))
    return results, (t1 - t0)

# ---------------------------
# Visualization helpers (new visuals)
# ---------------------------
def draw_boxes(image, detections):
    import matplotlib.patches as patches
    fig, ax = plt.subplots(1, figsize=(8,8))
    ax.imshow(image)
    for box, label, score in detections:
        x1,y1,x2,y2 = box
        rect = patches.Rectangle((x1,y1), x2-x1, y2-y1, linewidth=2, edgecolor='lime', facecolor='none')
        ax.add_patch(rect)
        ax.text(x1, y1-8, f"{label} {score:.2f}", color='yellow', fontsize=10, bbox=dict(facecolor='black', alpha=0.5))
    plt.axis('off')
    plt.show()

def plot_object_distribution(detections, title="Object Distribution"):
    labels = [d[1] for d in detections]
    counts = Counter(labels)
    if not counts:
        print("No objects to plot for distribution.")
        return
    items = sorted(counts.items(), key=lambda x: -x[1])
    keys = [k for k,_ in items]; vals = [v for _,v in items]
    plt.figure(figsize=(8,4))
    plt.bar(keys, vals)
    plt.xticks(rotation=45, ha='right')
    plt.title(title)
    plt.tight_layout()
    plt.show()

def plot_confidence_histogram(detections, title="Detection Confidence Histogram"):
    scores = [d[2] for d in detections]
    if not scores:
        print("No detections to plot confidences.")
        return
    plt.figure(figsize=(6,4))
    plt.hist(scores, bins=6, range=(0,1))
    plt.title(title)
    plt.xlabel("Confidence")
    plt.ylabel("Frequency")
    plt.show()

def plot_box_size_violin(detections, title="Bounding Box Size Distribution (violin)"):
    sizes = []
    for box, _, _ in detections:
        w = max(0, box[2]-box[0]); h = max(0, box[3]-box[1])
        sizes.append(w*h)
    if not sizes:
        print("No box sizes to plot.")
        return
    plt.figure(figsize=(6,4))
    plt.violinplot(sizes, showmeans=True)
    plt.title(title)
    plt.ylabel("Area (pixels)")
    plt.show()

def plot_cooccurrence_matrix(detections, title="Object Co-occurrence Matrix"):
    labels = [d[1] for d in detections]
    if len(labels) < 2:
        print("Not enough objects to compute co-occurrence.")
        return
    unique = sorted(list(set(labels)))
    idx = {k:i for i,k in enumerate(unique)}
    mat = np.zeros((len(unique), len(unique)), dtype=int)
    for i in range(len(labels)):
        for j in range(i+1, len(labels)):
            a = idx[labels[i]]; b = idx[labels[j]]
            mat[a,b] += 1
            mat[b,a] += 1
    plt.figure(figsize=(6,5))
    plt.imshow(mat, cmap='Blues')
    plt.colorbar()
    plt.xticks(range(len(unique)), unique, rotation=45, ha='right')
    plt.yticks(range(len(unique)), unique)
    plt.title(title)
    plt.tight_layout()
    plt.show()

def plot_caption_word_frequency(caption, top_k=20, title="Caption Word Frequency"):
    if not caption:
        print("No caption to analyze.")
        return
    words = [w.strip(".,!?;:()[]\"'").lower() for w in caption.split() if w.strip()]
    freq = Counter(words)
    if not freq:
        print("Caption tokenization produced no words.")
        return
    common = freq.most_common(top_k)
    keys = [k for k,_ in common]; vals = [v for _,v in common]
    plt.figure(figsize=(8,4))
    plt.bar(keys, vals)
    plt.xticks(rotation=45, ha='right')
    plt.title(title)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Benchmarking utilities (advanced)
# ---------------------------
def sample_system_usage():
    mem = psutil.virtual_memory().percent
    cpu = psutil.cpu_percent(interval=None)
    gpus = GPUtil.getGPUs() if torch.cuda.is_available() else []
    gpu_load = float(gpus[0].load*100) if gpus else 0.0
    return {"mem": mem, "cpu": cpu, "gpu": gpu_load, "ts": time.time()}

# run multiple inferences (with warmup) and collect timings & system metrics
def benchmark_images(images, repeats=REPEATS_PER_IMAGE, warmup=WARMUP_RUNS, threshold=DETECTION_THRESHOLD):
    # warm-up
    for _ in range(warmup):
        for img in images:
            try:
                _ = generate_caption(img)
                _ = run_detection(img, threshold=threshold)
            except Exception:
                pass

    per_image_stats = []  # list of dicts for each image
    system_samples = []   # aggregated time-series samples across runs

    # iterate images with progress bar
    for img in tqdm(images, desc="Benchmarking images"):
        stats = {"caption_times": [], "detection_times": [], "total_times": [], "system_snapshots": []}
        for r in range(repeats):
            t0 = time.time()
            sys_before = sample_system_usage()
            try:
                cap, tcap = generate_caption(img)
            except Exception as e:
                cap, tcap = ("error", None)
            sys_mid = sample_system_usage()
            try:
                dets, tdet = run_detection(img, threshold=threshold)
            except Exception as e:
                dets, tdet = ([], None)
            t1 = time.time()
            stats["caption_times"].append(tcap if tcap is not None else np.nan)
            stats["detection_times"].append(tdet if tdet is not None else np.nan)
            stats["total_times"].append(t1 - t0)
            stats["system_snapshots"].append((sys_before, sys_mid))
            system_samples.append((sys_before, sys_mid))
        per_image_stats.append(stats)
    return per_image_stats, system_samples

# ---------------------------
# PDF Report generation
# ---------------------------
def generate_pdf_report(report_path, metadata, figures_buffered):
    """
    figures_buffered: list of (figure_title, matplotlib_figure_callable)
    We'll save all provided figures into a single PDF and include metadata page as plaintext.
    """
    try:
        with PdfPages(report_path) as pdf:
            # metadata page
            fig = plt.figure(figsize=(11,8.5))
            plt.axis('off')
            meta_text = json.dumps(metadata, indent=2)
            plt.text(0.01, 0.99, "Model Documentation Report", fontsize=18, weight='bold', va='top')
            plt.text(0.01, 0.95, f"Generated: {now_str()} (UTC)", fontsize=10, va='top')
            plt.text(0.01, 0.9, "Metadata (JSON):", fontsize=12, va='top')
            # wrap meta text
            plt.text(0.01, 0.86, meta_text, fontsize=8, va='top', family='monospace')
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)

            # save all figures by invoking their callables
            for title, fig_callable in figures_buffered:
                try:
                    fig = fig_callable()  # the function should create and return the current figure
                    if fig is None:
                        # some callables may directly plot; capture current figure
                        fig = plt.gcf()
                    pdf.savefig(fig, bbox_inches='tight')
                    plt.close(fig)
                except Exception as e:
                    print("Failed to add figure to PDF:", title, e)
        print("Saved PDF report to", report_path)
    except Exception as e:
        print("Error generating PDF:", e)

# Helper wrappers to capture figures for PDF: each wrapper returns a callable that when called plots and returns figure
def fig_wrapper_plot_object_distribution(detections, title="Object Distribution"):
    def fn():
        plt.figure(figsize=(8,4))
        labels = [d[1] for d in detections]
        counts = Counter(labels)
        items = sorted(counts.items(), key=lambda x: -x[1])
        if not items:
            plt.text(0.5,0.5,"No objects detected", ha='center')
        else:
            keys = [k for k,_ in items]; vals = [v for _,v in items]
            plt.bar(keys, vals)
            plt.xticks(rotation=45, ha='right')
            plt.title(title)
            plt.tight_layout()
        return plt.gcf()
    return fn

def fig_wrapper_confidence_hist(detections, title="Detection Confidence Histogram"):
    def fn():
        plt.figure(figsize=(6,4))
        scores = [d[2] for d in detections]
        if not scores:
            plt.text(0.5,0.5,"No detection scores", ha='center')
        else:
            plt.hist(scores, bins=6, range=(0,1))
            plt.title(title)
            plt.xlabel("Confidence")
            plt.ylabel("Frequency")
        return plt.gcf()
    return fn

def fig_wrapper_heatmap(image, heatmap, title="Attention Heatmap (edge-based)"):
    def fn():
        fig, ax = plt.subplots(1, figsize=(8,8))
        ax.imshow(overlay_heatmap_on_image(image, heatmap))
        ax.set_title(title)
        ax.axis('off')
        return fig
    return fn

def fig_wrapper_detection_overlay(image, detections, title="Detections Overlay"):
    def fn():
        fig, ax = plt.subplots(1, figsize=(8,8))
        ax.imshow(image)
        import matplotlib.patches as patches
        for box, label, score in detections:
            x1,y1,x2,y2 = box
            rect = patches.Rectangle((x1,y1), x2-x1, y2-y1, linewidth=2, edgecolor='lime', facecolor='none')
            ax.add_patch(rect)
            ax.text(x1, y1-8, f"{label} {score:.2f}", color='yellow', fontsize=10, bbox=dict(facecolor='black', alpha=0.5))
        ax.set_title(title)
        ax.axis('off')
        return fig
    return fn

# ---------------------------
# Main: upload -> process -> visuals -> benchmark -> report
# (keeps original workflow)
# ---------------------------
print("\nPlease upload image(s) — Colab will prompt you.")
uploaded = files.upload()

# load images list (PIL) in order of upload
images = []
filenames = []
for fname in uploaded.keys():
    try:
        img = Image.open(fname).convert("RGB")
        images.append(img)
        filenames.append(fname)
    except Exception as e:
        print("Failed to open uploaded file:", fname, e)

# If no models loaded, still proceed but warn
if blip_model is None:
    print("Warning: BLIP model not loaded — captions will be skipped.")
if det_model is None:
    print("Warning: Detection model not loaded — detections will be skipped.")

# Advanced benchmarking (timings + system usage)
per_image_stats, system_samples = benchmark_images(images, repeats=REPEATS_PER_IMAGE, warmup=WARMUP_RUNS)

# Per-image results and visualizations (with progress bar)
report_metadata = {
    "generated_at": now_str(),
    "device": DEVICE,
    "detection_threshold": DETECTION_THRESHOLD,
    "files": [],
    "benchmark_summary": {}
}
figures_for_pdf = []

for i, img in enumerate(tqdm(images, desc="Processing uploaded images")):
    fname = filenames[i]
    entry = {"file": fname, "caption": None, "detections": None, "timings": per_image_stats[i]}
    # Attempt caption + detection (we keep same workflow order)
    try:
        caption, tcap = generate_caption(img)
        entry["caption"] = {"text": caption, "time": tcap}
    except Exception as e:
        caption = "ERROR_GENERATING_CAPTION"
        entry["caption"] = {"text": caption, "time": None, "error": str(e)}
        print("Caption error for", fname, e)

    try:
        detections, tdet = run_detection(img)
        entry["detections"] = {"count": len(detections), "time": tdet, "items": [{"label":d[1],"score":d[2],"bbox":[float(v) for v in d[0]]} for d in detections]}
    except Exception as e:
        detections = []
        entry["detections"] = {"count": 0, "time": None, "items": [], "error": str(e)}
        print("Detection error for", fname, e)

    # Show original
    print(f"\n--- {fname} ---")
    show_image(img, title=f"Uploaded: {fname}")

    # Caption printing
    print("\n📌 CAPTION:")
    print(entry["caption"]["text"])

    # Detections printed and drawn
    print("\n📌 DETECTIONS:")
    for d in detections:
        print(f" - {d[1]} ({d[2]:.2f})  bbox={list(map(int,d[0]))}")
    if detections:
        draw_boxes(img, detections)

    # Heatmap (edge-based) and show
    gray = np.mean(np.array(img), axis=2)
    heatmap = sobel(gray)
    show_image(overlay_heatmap_on_image(img, heatmap), title="Attention Heatmap (edge-based)")

    # New visualizations
    plot_object_distribution(detections)
    plot_confidence_histogram(detections)
    plot_box_size_violin(detections)
    plot_cooccurrence_matrix(detections)
    plot_caption_word_frequency(entry["caption"]["text"])

    # Add figures for PDF (wrap important ones)
    figures_for_pdf.append(("Detections Overlay - " + fname, fig_wrapper_detection_overlay(img, detections)))
    figures_for_pdf.append(("Heatmap - " + fname, fig_wrapper_heatmap(img, heatmap)))
    figures_for_pdf.append(("Object Distribution - " + fname, fig_wrapper_plot_object_distribution(detections)))
    figures_for_pdf.append(("Confidence Histogram - " + fname, fig_wrapper_confidence_hist(detections)))

    # Save metadata entry
    report_metadata["files"].append(entry)

# Summarize benchmarking stats (compute percentiles)
all_caption_times = []
all_detection_times = []
all_total_times = []
for stats in per_image_stats:
    all_caption_times.extend([t for t in stats["caption_times"] if t is not None])
    all_detection_times.extend([t for t in stats["detection_times"] if t is not None])
    all_total_times.extend([t for t in stats["total_times"] if t is not None])

def percentile(arr, p):
    if not arr:
        return None
    return float(np.percentile(np.array(arr), p))

report_metadata["benchmark_summary"] = {
    "caption_mean": float(np.nanmean(all_caption_times)) if all_caption_times else None,
    "detection_mean": float(np.nanmean(all_detection_times)) if all_detection_times else None,
    "total_mean": float(np.nanmean(all_total_times)) if all_total_times else None,
    "caption_50p": percentile(all_caption_times, 50),
    "caption_90p": percentile(all_caption_times, 90),
    "detection_50p": percentile(all_detection_times, 50),
    "detection_90p": percentile(all_detection_times, 90),
    "total_50p": percentile(all_total_times, 50),
    "total_90p": percentile(all_total_times, 90),
    "system_samples_count": len(system_samples)
}

# Save JSON metadata
json_path = os.path.join(OUTPUT_DIR, f"report_metadata_{now_str()}.json")
with open(json_path, "w", encoding="utf-8") as jf:
    json.dump(report_metadata, jf, indent=2)
print("\nSaved metadata JSON to", json_path)

# Generate PDF report (figures + metadata)
pdf_path = os.path.join(OUTPUT_DIR, f"model_report_{now_str()}.pdf")
generate_pdf_report(pdf_path, report_metadata, figures_for_pdf)

print("\nAll done. Outputs saved to:", OUTPUT_DIR)
print("Files:", os.listdir(OUTPUT_DIR))
print("Download the PDF from the Colab file browser or use the Colab `files` utilities to fetch it.")
```

```python
def generate_caption(image):
    if image is None:
        return "Please upload an image."

    image = image.convert("RGB")
    inputs = processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        out = model.generate(**inputs, max_length=60)

    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption
```

```python
def chatbot_reply(image, user_message):
    caption = generate_caption(image)

    if user_message:
        reply = f"""
🖼️ Image Description:
{caption}

💬 Your Question:
{user_message}
"""
    else:
        reply = f"""
🖼️ Image Description:
{caption}
"""

    return reply.strip()
```

```python
import torch
import torchvision
from torchvision import transforms as T
from PIL import Image, ImageDraw
import gradio as gr
from collections import Counter

from transformers import BlipProcessor, BlipForConditionalGeneration

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

```python
print("Loading BLIP caption model...")
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
caption_model = BlipForConditionalGeneration.from_pretrained(
    "Salesforce/blip-image-captioning-base"
).to(device)
caption_model.eval()

print("Loading Faster R-CNN detection model...")
det_model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
det_model = det_model.to(device)
det_model.eval()

print("Models loaded successfully!")
```

```python
COCO_CLASSES = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'N/A', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'N/A', 'backpack', 'umbrella',
    'N/A', 'N/A', 'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard',
    'sports ball', 'kite', 'baseball bat', 'baseball glove', 'skateboard',
    'surfboard', 'tennis racket', 'bottle', 'N/A', 'wine glass', 'cup', 'fork',
    'knife', 'spoon', 'bowl', 'banana', 'apple', 'sandwich', 'orange', 'broccoli',
    'carrot', 'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch', 'potted plant',
    'bed', 'N/A', 'dining table', 'N/A', 'N/A', 'toilet', 'N/A', 'tv', 'laptop',
    'mouse', 'remote', 'keyboard', 'cell phone', 'microwave', 'oven', 'toaster',
    'sink', 'refrigerator', 'N/A', 'book', 'clock', 'vase', 'scissors', 'teddy bear',
    'hair drier', 'toothbrush'
]

to_tensor = T.Compose([T.ToTensor()])
```

```python
def generate_caption(image):
    image = image.convert("RGB")
    inputs = processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        out = caption_model.generate(**inputs, max_length=60)

    return processor.decode(out[0], skip_special_tokens=True)
```

```python
def detect_and_draw(image, threshold=0.5):
    img_t = to_tensor(image).to(device)

    with torch.no_grad():
        outputs = det_model([img_t])[0]

    boxes = outputs["boxes"].cpu().numpy()
    labels = outputs["labels"].cpu().numpy()
    scores = outputs["scores"].cpu().numpy()

    draw = ImageDraw.Draw(image)
    results = []

    for box, lab, score in zip(boxes, labels, scores):
        if score >= threshold:
            name = COCO_CLASSES[lab] if lab < len(COCO_CLASSES) else f"label_{lab}"
            results.append(name)

            x1, y1, x2, y2 = box
            draw.rectangle([x1, y1, x2, y2], outline="lime", width=3)
            draw.text((x1, y1 - 10), f"{name} {score:.2f}", fill="yellow")

    return image, results
```
